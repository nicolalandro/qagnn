Predict

```python
import numpy as np
import torch
from tqdm import tqdm

from modeling.modeling_encoder import MODEL_NAME_TO_CLASS
from modeling.modeling_qagnn import LM_QAGNN, LM_QAGNN_DataLoader
from utils.data_utils import MultiGPUSparseAdjDataBatchGenerator, load_input_tensors, \
    load_sparse_adj_data_with_contextnode
```

```python
args = type('test', (object,), {})()

args.model_path = 'saved_models/csqa_model_hf3.4.0.pt'
args.ent_emb_paths = ['data/cpnet/tzw.ent.npy']
args.dev_statements = output_paths['statement']
args.dev_adj = output_paths['graph']

args.batch_size = 1
args.eval_batch_size = 1
args.subsample = 1.0
args.use_cache = True

device0 = torch.device("cpu")
device1 = torch.device("cpu")
```

```python
cp_emb = [np.load(path) for path in args.ent_emb_paths]
cp_emb = torch.tensor(np.concatenate(cp_emb, 1), dtype=torch.float)
concept_num, concept_dim = cp_emb.size(0), cp_emb.size(1)
print('| num_concepts: {} |'.format(concept_num))
```

```python
model_state_dict, old_args = torch.load(args.model_path, map_location=torch.device('cpu'))
model = LM_QAGNN(old_args, old_args.encoder, k=old_args.k, n_ntype=4, n_etype=old_args.num_relation,
                 n_concept=concept_num,
                 concept_dim=old_args.gnn_dim,
                 concept_in_dim=concept_dim,
                 n_attention_head=old_args.att_head_num, fc_dim=old_args.fc_dim, n_fc_layer=old_args.fc_layer_num,
                 p_emb=old_args.dropouti, p_gnn=old_args.dropoutg, p_fc=old_args.dropoutf,
                 pretrained_concept_emb=cp_emb, freeze_ent_emb=old_args.freeze_ent_emb,
                 init_range=old_args.init_range,
                 encoder_config={})
model.load_state_dict(model_state_dict)
model.encoder.to(device0)
model.decoder.to(device1)
model.eval()
```

```python
# deallocate RAM
cp_emb = None
```

```python
max_seq_length = 128
model_name = old_args.encoder
model_type = MODEL_NAME_TO_CLASS[model_name]
dev_qids, dev_labels, *dev_encoder_data = load_input_tensors(args.dev_statements, model_type, model_name,
                                                             max_seq_length)

num_choice = 5
max_node_num=200
*dev_decoder_data, dev_adj_data = load_sparse_adj_data_with_contextnode(args.dev_adj, max_node_num, num_choice, args)

dev_dataset = MultiGPUSparseAdjDataBatchGenerator(
    args, 'eval', device0, device1, args.eval_batch_size,
    torch.arange(len(dev_qids)), dev_qids, dev_labels,
    tensors0=dev_encoder_data,
    tensors1=dev_decoder_data,
    adj_data=dev_adj_data
)
```

```python
n_samples, n_correct = 0, 0

with torch.no_grad():
    for qids, labels, *input_data in tqdm(dev_dataset):
        logits, _ = model(*input_data)
        print(logits.argmax(1))
        n_correct += (logits.argmax(1) == labels).sum().item()
        n_samples += labels.size(0)
        print(f'the response is:', logits.argmax(1).item())
        break
print('Accuracy:', n_correct / n_samples)
```
